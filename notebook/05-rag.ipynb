```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model(
    model="openai:gpt-4o",
    temperature=0
)
```

```python
file_search_tool = {
    "type": "file_search",
    "vector_store_ids": ["vs_691b86cb24788191ad17cfdc400f3e4a"],
    # "description": "Search for files in the file system",
    # "args": {
    #     "path": {
    #         "type": "string",
    #         "description": "Path to search for files"
    #     }
    # }
}
llm_with_tools = llm.bind_tools([file_search_tool])
```

```python
response = llm_with_tools.invoke("Cómo mejoro el rendimiento de mi website?")
response.pretty_print()
```
