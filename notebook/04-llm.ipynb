```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

gpt4o_llm = ChatOpenAI(model_name="gpt-4o", temperature=0)
response = gpt4o_llm.invoke("Hola cómo estás?")
response.pretty_print()
```

```python
gpt3_llm = ChatOpenAI(model_name="gpt-3.5-turbo-0125", temperature=0)
response = gpt3_llm.invoke("Hola cómo estás?")
response.pretty_print()
```

```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

msg_1 = SystemMessage(content="You are a helpful assistant that talks in spanish.")
msg_2 = HumanMessage(content="Me llamo Osvaldo")
msg_3 = AIMessage(content="Hola Osvaldo, ¿en qué puedo ayudarte?")
msg_4 = HumanMessage(content="Cómo me llamo?")

history = [msg_1, msg_2, msg_3, msg_4]
```

```python
response = gpt3_llm.invoke(history)
response.pretty_print()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

gemini_llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0,
)

response = gemini_llm.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model(
    model="google_genai:gemini-2.5-flash",
    temperature=0
)

response = llm.invoke(history)
response.pretty_print()
```
